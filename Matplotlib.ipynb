```python
import numpy as np
X = np.random.randint(9,size= 11)
Y = X ** 2

plt.plot(X, Y, 'b') # 'b' is the color blue
plt.xlabel('X axis title')
plt.ylabel('Y axis title')
plt.title('Heading')
```

```python
A= np.random.normal(0,10,1000)
B= A/9
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(A, B, 'r--') # More on color options later
plt.subplot(1,2,2)
plt.plot(B, A, 'b*-');
```

# Scatter plot 


```python
## genera random sample data

sample = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50)}
sample['b'] = sample['a'] + np.random.randn(50)* 5



plt.scatter('a', 'b', c='c', data=sample) # c is the color
plt.xlabel('a axis')
plt.ylabel('b axis')
```

```python
import numpy as np
x = np.linspace(0, 10, 2000) # Return evenly spaced numbers over a specified interval.
fig = plt.figure(figsize=(18,5))# to specify the size of the width and height of the figure in inches
plt.plot(x, np.sin(x), '--', c = 'r') ## 
plt.plot(x, np.cos(x), '*-', c='g')
```

##### 2D array

```python
n = np.arange(15)

fig, axes = plt.subplots(1, 4, figsize=(15,5))

axes[0].scatter(n, n ** 3, color = "b")
axes[0].set_title("scatter")

axes[1].step(n, n**2, lw=2, color = "y")
axes[1].set_title("step")

axes[2].bar(n, n**2,color="purple")
axes[2].set_title("bar")

axes[3].fill_between(n, n**2, n**3, color="green", alpha=0.5);
axes[3].set_title("fill_between");
```
